Thus, the review rating has 37 missing values. Here, among mean and median, most of the experts prefer median over mean because it considers outliers. Median should be used when data has outliers and mean should be used when data is normally distributed and has no extreme outliers.

here, the review rating of food, clothing item differs significantly. So, first we calculate the median rating of all and then only fill in those missing values.


```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df.isnull().sum()
```

```python
# 2. Now clean it properly for Lowercase
df.columns = df.columns.str.lower()
```

```python
df.columns = df.columns.str.replace(' ', '_')
```

```python
df.columns
```

```python
df = df.rename(columns={'purchase_amount_(usd)': 'purchase_amount'})
```

```python
df.columns
```

```python
## Feature engineering on age column by creating age group
bins = [0, 18, 25, 35, 50, 100]
labels = ['Teen', 'Young Adult', 'Adult', 'Mid-Age', 'Senior']

df['age_group'] = pd.cut(df['age'], bins=bins, labels=labels)

### we can use cut by providing bins and qcut without providing bins
```

```python
df[['age','age_group']].head()
```

```python
df['frequency_of_purchases'].unique()
```

Here, we have noticed that fornightly and bi-weekly are same, and every 3 months and quarterly has same values.

```python
#Create column frequency of purchases in days using map()
frequency_map = {
    'Weekly': 7,
    'Bi-weekly': 14,
    'Fortnightly': 14,
    'Every 3 Months': 90,
    'Monthly': 30,
    'Quarterly': 90,
    'Annually': 365
}
df['purchase_frequency_in_days'] = df['frequency_of_purchases'].map(frequency_map)
```

```python
df[['frequency_of_purchases','purchase_frequency_in_days']].head(10)
```

```python
df[['discount_applied','promo_code_used']].head()
```

Here, it looks like both column have same values. If promo code is used, then obviously, the customer will get discount. Lets check it.

```python
(df['discount_applied']==df['promo_code_used']).all()
```

Since both column has same values, we are going to remove promo_code column.

```python
df= df.drop('promo_code_used',axis=1)
```
